```python
# 1) Instalar dependencias y localizar el script
!pip install -q ultralytics opencv-python-headless
from pathlib import Path

candidates = list(Path('/kaggle/input').rglob('kaggle_train_powerai.py'))
if not candidates:
    raise FileNotFoundError('kaggle_train_powerai.py no encontrado. Subelo junto al .zip en el Dataset.')
SCRIPT = candidates[0]
print('Script encontrado:', SCRIPT)
```

```python
# 2) Entrenar
# 6175 imagenes etiquetadas, 2 clases: plate y bar_hub
MODE = 'train'
EPOCHS = 100       # dataset grande, converge antes
BATCH = 8          # baja a 4 si Kaggle se queda sin memoria

import subprocess
cmd = ['python', str(SCRIPT), '--mode', MODE, '--device', '0',
       '--detector-base-model', 'yolo11s.pt',
       '--detector-epochs', str(EPOCHS), '--detector-batch', str(BATCH)]
print(' '.join(cmd))
subprocess.run(cmd, check=True)
```
